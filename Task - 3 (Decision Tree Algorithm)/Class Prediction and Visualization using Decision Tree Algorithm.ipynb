### Predicting the right class for the 'Iris' data using Decision Tree Algorithm and visualizing it graphically 

```python
#Importing the libraries
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the iris dataset from sklearn.datasets
iris = datasets.load_iris()
```

```python
#Loading the iris dataset into pandas dataframe
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

Decision Tree Classifier:

The classification technique is a systematic approach to build classification models from an input data set. For example, Decision Tree Classifiers, Rule-Based Classifiers, Neural Networks, Support Vector Machines, and Naive Bayes Classifiers are different techniques to solve a classification problem. Each technique adopts a learning algorithm to identify a model that best fits the relationship between the attribute set and class label of the input data. Therefore, a key objective of the learning algorithm is to build a predictive model that accurately predict the class labels of previously unknown records.

Decision Tree Classifier is a simple and widely used Classification Technique. It applies a straightforward idea to solve the classification problems. Decision Tree Classifier poses a series of carefully crafted questions about the attributes of the test record.

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn import tree
plt.figure(figsize = (20,15), facecolor = 'white', dpi = 150)
dtree.fit(df, y)
tree_plt = tree.plot_tree(dtree, feature_names = iris.feature_names, fontsize = 13, precision = 2, filled = True, proportion = True, rounded = 10)
plt.show()
```
